#### Mini preprocessing before going to concrete questions:

```python
import copy
# Handling Missing Values
# Drop rows with missing values in 'PropType'
df.dropna(subset=['PropType'], inplace=True)

# Data Cleaning
# Remove rows with zero values for 'Year_Built' and 'Sale_price'
df = df[(df['Year_Built'] != 0) & (df['Sale_price'] != 0) & (df['Fin_sqft'] != 0)]

# Outlier Detection and Removal (using z-score)
from scipy.stats import zscore
z_scores = zscore(df[['Fin_sqft', 'Sale_price']])
df = df[(z_scores < 3).all(axis=1)]  # Keep rows where z-scores for 'Fin_sqft' and 'Sale_price' are less than 3

# Add a day value to the 'Sale_date' column (e.g., 01 for the first day of the month)
df['Sale_date'] = df['Sale_date'] + '-01'

# Convert 'Sale_date' to datetime format
df['Sale_date'] = pd.to_datetime(df['Sale_date'], format='%Y-%m-%d')


# Adjust prices by taking into account inflation 2002-2018 (https://www.officialdata.org/us/inflation/2002?endYear=2018&amount=100)
inflations = [1.58, 2.28, 2.66, 3.39, 3.23, 2.85, 3.84, -0.36, 1.64, 3.16, 2.07, 1.46, 1.62, 0.12, 1.26, 2.13, 2.49]
def calculate_adjustment(row):
    year = row['Sale_date'].year
    inflation_rates = inflations[year - 2002:2019 - 2002]  # Adjust for the range from the sale year to 2018
    price = copy.copy(row['Sale_price'])
    for inf_rate in inflation_rates:
         price = price * (1 + inf_rate / 100)
    return price

# Apply the function to each row to calculate the inflation-adjusted price
df['Inflation_adjusted_price'] = df.apply(calculate_adjustment, axis=1)

# Calculate price per square meter
df['Price_per_sqm'] = df['Inflation_adjusted_price'] / (df['Fin_sqft'] * 0.092903)

df.info()
```

### Kolmogorov-Smirnov Test
After applying `np.log1p`, let us firstly test the following hypotheses:

$H0$: a given sample of log sale prices follows normal distribution

$H1$: a given sample does not follow normal distribution

```python
from scipy.stats import kstest

# Calculate mean and standart deviation
mean = np.mean(log_residential_prices)
std = np.std(log_residential_prices)

# perform Kolmogorov-Smirnov test
_, p_value = kstest(log_residential_prices, 'norm', args=(mean, std))

if p_value < 0.05 or p_value is None:
  print("p-value is equal to {}, so we can reject the null hypothesis".format(p_value))
else:
  print("p-value is equal to {}, so we cannot reject the null hypothesis".format(p_value))
```

```python
mean = np.mean(log_commercial_prices)
std = np.std(log_commercial_prices)

_, p_value = kstest(log_commercial_prices,
                    'norm',
                    args=(mean, std))

p_value
```

```python
mean = np.mean(log_residential_prices)
std = np.std(log_residential_prices)

_, p_value = kstest(log_residential_prices,
                    'norm',
                    args=(mean, std))

p_value
```

> The observed disparity in variances between the two populations precludes the application of the conventional T-test methodology.

> Additionally, preliminary analysis involving visualization via plots was conducted prior to the application of the logarithmic transformation (log1p function). Since we are dealing with data that has a lot of outliers. This transformation can help make the distribution more symmetric and reduce the impact of extreme values.

> Using Kolmogorov-Smirnov Test we also rejected hypothesis, that logarithm of sale prices per sqm is normally distributed.

> Consequently, the Mann-Whitney U test emerges as the preferred statistical approach for comparative analysis in this scenario.

### Mann-Whitney U test

#### Null Hypothesis ($H_0$):
- There is no significant difference in the sale prices between residential and commercial properties.

#### Alternative Hypothesis ($H_1$):
- There is a significant difference in the sale prices between residential and commercial properties.

> By using the Mann-Whitney U test, we can make a comparison between the distributions of sale prices for residential and commercial properties without relying on the assumption of normality. This ensures that the results are robust and reliable, even in the presence of non-normally distributed data.


```python
from scipy.stats import mannwhitneyu

result = mannwhitneyu(log_residential_prices,
                      log_commercial_prices)

U_test = result.statistic
result.pvalue
```

**Reject null hypothesis:**

There is a significant difference in sale prices between residential and commercial properties.


#### Mann-Whitney-U-Test Effect Size

[How to calculate](https://en.wikipedia.org/wiki/Mann%E2%80%93Whitney_U_test#Rank-biserial_correlation
)



```python
n_r = len(df_filtered.loc[df_filtered['PropType'] == 'Residential'])
n_c = len(df_filtered.loc[df_filtered['PropType'] == 'Commercial'])
```

```python
effect_size = 1 - (2*U_test)/(n_r*n_c)

effect_size
```

The utilization of the Rank-biserial correlation as a measure of effect size in the context of the Mann-Whitney U test for comparing sale prices between residential and commercial properties in Milwaukee is a methodological choice grounded in its appropriateness for assessing the magnitude of the observed difference between the two groups.

The Rank-biserial correlation offers several advantages in this scenario:

1. **Sensitivity to Differences:** The Rank-biserial correlation captures the strength and direction of the relationship between the ranked sale prices and the grouping variable (residential vs. commercial properties).

2. **Robustness to Non-Normality:** Given that the Mann-Whitney U test is a non-parametric test used when the assumption of normality is violated, the Rank-biserial correlation complements this by offering a non-parametric measure of effect size that is robust to deviations from normality in the data distribution. This ensures the validity of the effect size estimation in the presence of non-normally distributed sale prices.

3. **Interpretability:** The Rank-biserial correlation coefficient ranges from -1 to 1, where values closer to 1 indicate a stronger positive relationship between the variables (i.e., higher sale prices for residential properties compared to commercial properties), while values closer to -1 suggest a stronger negative relationship (i.e., higher sale prices for commercial properties compared to residential properties). A value of 0 indicates no association between the variables. In this context, a Rank-biserial correlation of 0.3729846355009364 suggests a moderate effect size, indicating a meaningful difference in sale prices between residential and commercial properties in Milwaukee.

4. **Comparability:** The use of the Rank-biserial correlation facilitates comparability with effect sizes reported in other studies utilizing non-parametric tests, allowing for cross-study comparisons and meta-analytic synthesis of findings.

Given the moderate effect size observed in this analysis, it is reasonable to conclude that there exists a meaningful and statistically significant difference in sale prices between residential and commercial properties in Milwaukee. This finding underscores the importance of considering property type as a significant determinant of sale price variability in the real estate market.

```python
from numpy import sqrt

# Calculate Cohen's d for residential properties
mean_diff_residential = crisis_residential_prices['Price_per_sqm'].mean() - stable_residential_prices['Price_per_sqm'].mean()
std_pooled_residential = sqrt((crisis_residential_prices['Price_per_sqm'].std() ** 2 + stable_residential_prices['Price_per_sqm'].std() ** 2) / 2)
cohen_d_residential = mean_diff_residential / std_pooled_residential

# Calculate Cohen's d for commercial properties
mean_diff_commercial = crisis_commercial_prices['Price_per_sqm'].mean() - stable_commercial_prices['Price_per_sqm'].mean()
std_pooled_commercial = sqrt((crisis_commercial_prices['Price_per_sqm'].std() ** 2 + stable_commercial_prices['Price_per_sqm'].std() ** 2) / 2)
cohen_d_commercial = mean_diff_commercial / std_pooled_commercial

print("Cohen's d for Residential Properties:", cohen_d_residential)
print("Cohen's d for Commercial Properties:", cohen_d_commercial)
```

### Kolmogorov-Smirnov test

> To determine whether the sale price varies across different exterior wall materials, we will indeed use the Kruskal-Wallis test, which is a non-parametric alternative to one-way ANOVA. Before conducting the test, let's assess the assumptions and ensure the data meets the requirements:

**Normality Assumption:**
- Since the Kruskal-Wallis test is non-parametric, it does not assume normality. However, it's still useful to check the distribution of the data using the Kolmogorov-Smirnov (KS) test for each group.

```python
# Check Normality using KS Test
alpha = 0.05

print("KS for extwall styles:")
for group, data in df.groupby('Extwall'):
    mean = np.mean(data['Price_per_sqm'])
    std = np.std(data['Price_per_sqm'])
    p_value = kstest(data['Price_per_sqm'], 'norm', args=(mean, std)).pvalue
    if p_value < alpha:
        print(f"Data for '{group}' does not follow a normal distribution (KS Test p-value = {p_value})")
    else:
        print(f"Data for '{group}' follows a normal distribution (KS Test p-value = {p_value})")
```

### Kruskal-Wallis Test

Hypothesis Testing:

Null Hypothesis ($H_0$): There is no significant difference in the median sale prices across different exterior wall materials

Alternative Hypothesis ($H_1$): There is a significant difference in the median sale prices across the different ex. wall materials

```python
from scipy.stats import kruskal

H_extwall, p_value_extwall = kruskal(*[group[1]['Price_per_sqm']
                                       for group in df.groupby('Extwall')])

print(f"H = {H_extwall}, p-value = {p_value_extwall}")
```

The Kruskal-Wallis test results indicate significant differences in sale prices across different exterior wall materials. Here's how we interpret these results:

  - p-value ≈ $9.09 \times 10^{-260}$ (extremely small)

- **Interpretation:**
  - With an extremely small p-value, we reject the null hypothesis.
  - Thus, we conclude that there is a significant difference in median sale prices across different exterior wall materials.

These results provide statistical evidence that exterior wall materials have a significant impact on sale prices.

### More insight on different Exterior Wall Materials:

#### Post-hoc Testing:
We will conduct post-hoc tests to identify significant pairwise differences in median sale prices among different exterior wall materials. Since we have already performed the Kruskal-Wallis test, we can proceed with pairwise Mann-Whitney U tests with Bonferroni correction for multiple comparisons.

#### Effect Size Calculation:
We will calculate effect sizes to measure the strength of the observed differences in median sale prices between exterior wall materials. We can use Cohen's d as the effect size measure.


```python
from scipy.stats import mannwhitneyu

def cohens_d(group1, group2):
    """Calculate Cohen's d effect size."""
    mean_diff = np.mean(group1) - np.mean(group2)
    n1, n2 = len(group1), len(group2)
    var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)
    pooled_var = ((n1 - 1) * var1 + (n2 - 1) * var2) / (n1 + n2 - 2)
    d = mean_diff / np.sqrt(pooled_var)
    return d

# Post-hoc Testing: Pairwise Mann-Whitney U tests with Bonferroni correction
alpha = 0.05
extwall_types = df['Extwall'].unique()
significant_pairs = []

for i in range(len(extwall_types)):
    for j in range(i+1, len(extwall_types)):
        extwall1 = extwall_types[i]
        extwall2 = extwall_types[j]

        group1 = df[df['Extwall'] == extwall1]['Price_per_sqm']
        group2 = df[df['Extwall'] == extwall2]['Price_per_sqm']

        if len(group1) == 0 or len(group2) == 0:
            continue  # Skip this pair if either group has zero size

        # Perform Mann-Whitney U test
        stat, p_value = mannwhitneyu(group1, group2, alternative='two-sided')
        corrected_alpha = alpha / len(extwall_types)

        if p_value < corrected_alpha:
            significant_pairs.append((extwall1, extwall2, p_value))

# Interpret Pairwise Comparisons
print("Significant Pairwise Comparisons:")
for pair in significant_pairs:
    print(f"- '{pair[0]}' vs '{pair[1]}; p-value: {pair[2]}'")

print()
print("Effect Size Calculation: Cohen's d")

# Effect Size Calculation: Cohen's d
effect_sizes_extwall = {}
for extwall1 in extwall_types:
    for extwall2 in extwall_types:
        if extwall1 != extwall2:
            group1 = df[df['Extwall'] == extwall1]['Price_per_sqm']
            group2 = df[df['Extwall'] == extwall2]['Price_per_sqm']
            effect_size = cohens_d(group1, group2)
            effect_sizes_extwall[f"'{extwall1}' vs '{extwall2}'"] = effect_size

# Interpret Effect Sizes
for comparison, effect_size in effect_sizes_extwall.items():
    print(f"Effect Size ({comparison}): {effect_size:.2f}")
```
